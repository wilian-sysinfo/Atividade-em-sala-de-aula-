## 1. Blocos de código em Python

Em Python, blocos são definidos pela indentação.

```python
if True:
    print("Este comando está dentro do bloco")
    print("Este também está dentro")

print("Este está fora do bloco")
```

Observe que os comandos dentro do if estão indentados.

Normalmente usamos 4 espaços.

## 2. Estruturas Condicionais

Permitem que o programa tome decisões.

Estrutura básica:

```python
if condição:
    código
```

Exemplo

```python
idade = 20

if idade >= 18:
    print("Maior de idade")
```

Usando else

```python
idade = 16

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

Usando elif

```python
nota = 7

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

## 4. Laço while

O while executa enquanto uma condição for verdadeira.

```python
contador = 0

while contador < 5:
    print(contador)
    contador += 1
```

Exemplo

```python
numero = 1

while numero <= 10:
    print(numero)
    numero += 1
```

## 5. Controle de fluxo

**Break**

Interrompe o loop.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

**Continue**

Pula uma iteração.

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

## 7. Exemplo prático

**Exercício**

Calcule a média de uma lista de números.
Essa lista pode ter qualquer tamanho e qualquer número. Seu código tem que identificar o tamanho da lista automaticamente e calcular a média dos valores presentes.

```python
listando = [4,5,6,7,8,9,1,0]
soma = 0
for n in listando:
  soma += n           #somando o valor da lista
itens = len(listando) #quantidade de itens da lista metodo len()
media = soma / itens
print(media)
```

# **8. Desafio Final**

Crie um programa que:

* peça 5 números ao usuário

* guarde em uma lista

* mostre:

  --  maior número

  -- menor número

  -- média

Estrutura sugerida:

while

lista

for

```python
numeros = [] # lisata vazia

# input do usuario
for i in range(5):
    num = float(input(f"Digite o {i+1}º número: "))
    numeros.append(num)

# Calcula maior, menor e média
maior = max(numeros)                      #retorna o maior n da lista
menor = min(numeros)                      #retorna o menor n da lista
media = sum(numeros) / len(numeros)       #retorna a soma dos n da lista

# saida resultado
print("-" * 30)
print(f"Lista: {numeros}")
print(f"Maior número: {maior}")
print(f"Menor número: {menor}")
print(f"Média: {media:.2f}")
```
