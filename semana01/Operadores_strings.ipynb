```python
# sua empresa
lat = '-22.005320'
lon = '-47.891040'

# startup adquirida
latlon = '-22.005320;-47.891040'
```

Como podemos normalizar a forma com que as latitudes e longitudes são armazenadas para possam ser comparadas?

```python
#print( latlon.find(';'))
print(latlon.split(';'))
```

```python
separador = latlon.find(';')
lat_nova = latlon[:separador]
lon_nova = latlon[separador+1:]
print(lat_nova)
print(lon_nova)
```

## Comparadores lógicos

São resultados de comparações lógicas. Os operadores de comparação lógica são:

*   `>` (maior);
*   `<` (menor);
*   `==` (igual);
*   `>=` (maior ou igual);
*   `<=` (menor ou igual);
*   `!=` (diferente).


**Exemplo:** Caixa eletrônico

**Exemplo:** Cartão de crédito

```python
cod = '123'
usuario = '124'

if cod == usuario:
  print ('ACESO APROVADO')
else:
    print('Acesso Negado')
```
